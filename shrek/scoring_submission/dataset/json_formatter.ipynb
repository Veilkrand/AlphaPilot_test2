```python
import json
import os
import glob
import pprint

json_file = 'export-2019-03-07T08_54_11.788Z.json'

with open(json_file, 'r') as f:
    json_dict = json.load(f)
    
labels_dict = {}
testing_labels_dict = {}
confidence = 1.0

for dic in json_dict:
        
    geometry = dic["Label"]["inner_polygon"][0]["geometry"]
    
    polygon = []
    for j in range(len(geometry)):
        polygon.append(geometry[j]['x'])
        polygon.append(geometry[j]['y'])
    
    # Create a copy to store clean ground truth
    labels = polygon
    labels = [labels]
    labels_dict[dic['External ID']] = labels
    
    # Add confidence score (not in ground truth)
    polygon.append(confidence)
    polygon = [polygon]
    testing_labels_dict[dic['External ID']] = polygon
    

with open('labels_with_confidence_for_testing.json', 'w') as outfile:
    json.dump(testing_labels_dict, outfile)

with open('labels_testLeaderBoard_v1.json', 'w') as outfile2:
    json.dump(labels_dict, outfile2)
```
